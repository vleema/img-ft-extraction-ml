## Teste estatístico dos classificadores individuais

````python
# gambiarra
from scipy.stats import friedmanchisquare


knn_24 = dtree_results[["dataset", "split"]].merge(
    knn_results,
    on=["dataset", "split"],
    how="left"
)


individual_models = pd.DataFrame({
    "mlp": mlp_results["conf 4"],
    "knn": knn_24["10k"],
    "dtree": dtree_results["conf. 3"]
})

stat, p = friedmanchisquare(*[individual_models[col] for col in individual_models.columns])

print("p-value:", p)

if p < 0.05:
    posthoc = sp.posthoc_nemenyi_friedman(individual_models.values)
    posthoc.index = individual_models.columns
    posthoc.columns = individual_models.columns

    print(posthoc)
````
